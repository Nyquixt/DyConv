```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
LOG_DIR = 'logs'
EXT = '-b128-e90.txt'
```

```python
models = ['cc', 'dy', 'dyresA', 'dyresB']
labels = ['CondConv', 'Dy-CNN', 'DDSNet-A', 'DDSNet-B']
backbones = ['alexnet', 'resnet18', 'mobilenetv2']
dataset = 'cifar100'
kernels = '4'
```

```python
Xs = []
```

```python
for m in models:
    file_name = m + kernels + '_' + backbones[0] + '-' + dataset + EXT
    with open(os.path.join(LOG_DIR, file_name)) as f:
        lines = f.readlines()
        lines = lines[5:-2]
        x = []
        for l in lines:
            stripped = l.strip()
            splitted = stripped.split(',')
            val = float(splitted[-1])
            x.append(val)
    Xs.append(x)
```

```python
len(Xs)
```

```python
y = [x for x in range(1, 91)]
plt.figure(figsize=(5, 5))
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax2 = plt.axes([.65, .6, .2, .2])
colors = ['blue', 'red', 'green', 'purple']
for i, x in enumerate(Xs):
    ax1.plot(y, x, color=colors[i], label=labels[i])
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Validation Accuracy')
ax1.set_title('n=4 AlexNet on CIFAR-100')
ax1.legend()
for i, x in enumerate(Xs):
    ax2.plot(y[60:91], x[60:91], color=colors[i])
plt.setp(ax2, xticks=[x for x in range(60, 91, 10)], yticks=[x for x in range(59, 61, 2)])

plt.savefig('n4_alexnet-cifar.pdf')
```

```python
dataset = 'tiny'
kernels = '4'
Xs = []
for m in models:
    file_name = m + kernels + '_' + backbones[1] + '-' + dataset + EXT
    with open(os.path.join(LOG_DIR, file_name)) as f:
        lines = f.readlines()
        lines = lines[5:-2]
        x = []
        for l in lines:
            stripped = l.strip()
            splitted = stripped.split(',')
            val = float(splitted[-1])
            x.append(val)
    Xs.append(x)

y = [x for x in range(1, 91)]
plt.figure(figsize=(5, 5))
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax2 = plt.axes([.65, .6, .2, .2])
colors = ['blue', 'red', 'green', 'purple']
for i, x in enumerate(Xs):
    ax1.plot(y, x, color=colors[i], label=labels[i])
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Validation Accuracy')
ax1.set_title('n=4 ResNet18 on Tiny ImageNet')
ax1.legend()
for i, x in enumerate(Xs):
    ax2.plot(y[60:91], x[60:91], color=colors[i])
plt.setp(ax2, xticks=[x for x in range(60, 91, 10)], yticks=[x for x in range(62, 63, 2)])
plt.savefig('n4_resnet18-tiny.pdf')
```
